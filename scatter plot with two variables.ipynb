```python
#import libraries
#define dataframe and read csv
#import file/dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('sub-data.csv', low_memory=False)
df.head()

#define variables
x = df['DEBT']
y = df['CCBAL']

plt.scatter(x = x, y = y, s = 1); #codes and plots the scatter plots with the variable that were previously defined
plt.title('Amount Of Debt and Credit Card Balance In Households') #title scatter plot
plt.xlabel('Debt in Households (in 100s of millions)') #label x-axis
plt.ylabel('Credit Card Balance(in dollars)'); #label y-axis
```
